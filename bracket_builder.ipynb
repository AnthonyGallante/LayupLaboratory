```python
def round_of_games(team_list: list, n: int=20):
    winners, losers = [], []
    schools = []
    records = []
    analyses = []

    for i in range(0, len(team_list), 2):
        game = simulate_n_games(team_list[i], team_list[i+1], n)
        winners.append(teams[game['Winner']])
        losers.append(teams[game['Loser']])
        analyses.append(game['Analysis'])
        schools.append([game['Team 1'], game['Team 2']])
        records.append([game['Win Count 1'], game['Win Count 2']])
    
    return {'Winners': winners, 
            'Losers': losers, 
            'Analysis': analyses, 
            'Schools': schools,
            'Records': records
        }


def first_round(region: dict, n: int=20):
    first_round_teams = [teams[v] for v in midwest.values()]
    return round_of_games(first_round_teams, n=n)
```
